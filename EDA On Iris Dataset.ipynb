Exploratory Data Analysis on the Iris Dataset

```python
import pandas as pd
```

Objective <P>
The objective of this analysis is to evaluate whether Iris flower species can be distinguished using available morphological features and to assess the suitability of these features for classification modeling.</p>

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df = df.drop(columns='Id')
```

```python
df.describe()
```

The summary statistics indicate noticeable variation across feature ranges, suggesting that certain measurements may contribute more strongly to species differentiation than others.

```python
df['Species'].value_counts()
```

The dataset consists of 150 observations across three Iris species.<br>
All three species are equally represented in the dataset.<br>
This balanced class distribution reduces the risk of model bias toward any single species.

```python
df['Species'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = df, x = 'Species')
```

```python
sns.histplot(data=df, x="SepalLengthCm", hue='Species')
```

```python
sns.histplot(data=df, x="PetalLengthCm", kde=True)
```

```python
sns.histplot(data=df, x="SepalWidthCm", kde=True)
```

Sepal-related features show moderate skewness, while petal-related features exhibit clearer separation patterns.<br>
Kernel Density Estimation highlights distinct density peaks for petal measurements, suggesting stronger discriminative potential.

```python
sns.histplot(data=df, x="PetalWidthCm", kde=True)
```

```python
sns.boxplot(
    data=df,
    x="Species",
    y="PetalWidthCm"
)
```

```python
sns.violinplot(data=df, x="Species", y="PetalLengthCm")
```

Iris-setosa exhibits lower variability compared to the other species, while Iris-virginica shows wider dispersion.<br>
Overlapping interquartile ranges between Iris-versicolor and Iris-virginica indicate potential classification challenges.

```python
sns.scatterplot(
    data=df,
    x="SepalLengthCm",
    y="SepalWidthCm",
    hue='Species'
)
```

Sepal-based scatter plots reveal clear clustering for Iris-setosa, whereas significant overlap exists between Iris-versicolor and Iris-virginica.<br>
This suggests that sepal features alone may be insufficient for robust classification.

```python
sns.scatterplot(
    data=df,
    x="PetalLengthCm",
    y="PetalWidthCm",
    hue='Species'
)
```

Petal length and petal width demonstrate strong separation across species when analyzed jointly.<br>
These features are likely to contribute most significantly to classification model performance.

```python
sns.pairplot(
    df,
    hue="Species",
    diag_kind="kde"
)
```

The exploratory analysis indicates that while sepal features provide limited separability, petal measurements offer strong discriminatory power across Iris species.<br>
As a result, classification models incorporating petal features are expected to achieve higher predictive accuracy.
